```python
sanitation_df = pd.read_csv('/content/gdrive/MyDrive/Raw Dataset/basic sanitation.csv')
sanitation_df
```

```python
sanitation_df.columns
```

```python
required_columns = ['Location', 'Period', 'Dim1' , 'Value']
sanitation_df2 = sanitation_df[required_columns]

sanitation_df2.columns= ['Country', 'Year', 'Area', 'Sanitation']
sanitation_df2

```

```python
sanitation_df2[sanitation_df2.Country=="Canada"]
sanitation_df2 = sanitation_df2[sanitation_df2.Year<2016]
```

```python
sanitation_df2 = sanitation_df2[sanitation_df2.Area=="Total"]
```

```python
sanitation_df2.drop('Area', axis='columns', inplace=True)
```

```python
sanitation_df2.shape
```

```python
sanitation_df2.sort_values(by=['Country','Year'],ignore_index=True, inplace=True)
```

```python
sanitation_df3 = sanitation_df2.copy()
sanitation_df3.drop_duplicates(inplace=True)
sanitation_df3 = sanitation_df3.sort_values(by=['Country', 'Year'],ignore_index=True)
sanitation_df3
```

```python
sanitation_df3.to_csv("sanitation_data.csv")
```
