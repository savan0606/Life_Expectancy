```python
gdp_df = pd.read_csv('/content/gdrive/MyDrive/Raw Dataset/gdp.csv')
gdp_df.columns
```

```python
gdp_df = gdp_df.melt(id_vars=["Country Name"], 
        var_name="year", 
        value_name="Gdp")
```

```python
gdp_df.columns = ['Country', 'Year', 'GDP']
gdp_df
```

```python
gdp_df.Year = pd.to_numeric(gdp_df.Year)
```

```python
gdp_df = gdp_df[gdp_df.Year<2016]
```

```python
gdp_df = gdp_df.sort_values(by=['Country', 'Year'],ignore_index=True)
gdp_df
```
