```python
dict1 = dict()
for country in country_names:
  country_no = df[df.Country==country]['Country'].count()
  dict1[country] = country_no

sorted_dict = sorted(dict1.items(), key=lambda x:x[1])
# sorted_dict
```

```python
l = list()
for i in range(len(sorted_dict)):
  if sorted_dict[i][1] < 16:
    l.append(sorted_dict[i][0])
```

```python
df1 = df[~df.Country.isin(l)]
```

```python
rdf = pd.read_csv('/content/gdrive/MyDrive/Cleaned Dataset/required_data.csv')
```

```python
rdf.columns
```

```python
rdf.rename(columns = {'Avg. Years of Schooling':'Schooling', 'HDI (Income Composition of Resources)':'Income composition of resources'}, inplace = True)
```

```python
merged_df = pd.merge(df1, rdf, on=['Country','Year'], how='outer')
```

```python
merged_df.head()
```

```python
merged_df.columns
```

```python
merged_df['Schooling'] = merged_df['Schooling_x'].fillna(merged_df['Schooling_y'])
merged_df['Income composition of resources'] = merged_df['Income composition of resources_x'].fillna(merged_df['Income composition of resources_y'])
```

```python
merged_df.drop(['Unnamed: 0_x', 'Unnamed: 0_y', 'Income composition of resources_y', 'Income composition of resources_x', 'Schooling_x', 'Schooling_y'], axis=1, inplace=True)
```

```python
merged_df.columns
```

```python
for c in merged_df[merged_df.Schooling.isna()].Country.unique():
  merged_df.loc[merged_df[merged_df.Country==c].index,'Schooling'] = merged_df[merged_df.Country==c]['Schooling'].fillna(0)
  merged_df.loc[merged_df[merged_df.Country==c].index,'Income composition of resources'] = merged_df[merged_df.Country==c]['Income composition of resources'].fillna(0)
```

```python
alc_c = df1[df1['Alcohol'].isna()].Country.unique()
for c in alc_c:
  for b in df1.loc[df1[df1.Country==c].index, 'Alcohol'].isna():
    if b:
      df1.loc[df1[df1.Country==c].index, 'Alcohol'] = df1[df1.Country==c]['Alcohol'].fillna(method="ffill")
```
