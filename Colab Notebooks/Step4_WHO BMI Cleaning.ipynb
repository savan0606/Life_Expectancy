```python
import pandas as pd

df = pd.read_csv('/content/gdrive/MyDrive/Raw Dataset/bmi.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.Dim2.unique()
```

```python
required_columns = ['Location', 'Period', 'Dim1', 'Dim2', 'FactValueNumeric']
df2 = df[required_columns]
df2.columns = ['Country', 'Year', 'Gender', 'Age_group', 'bmi']
```

```python
df2 = df2[df2.Year>1999]
df2 = df2[df2.Year<2016]
```

```python
df2 = df2.sort_values(by=['Country', 'Year'],ignore_index=True)
df2
```

```python
df2.Age_group.unique()
```

```python
df2 = df2[df2.Age_group!='5-19  years']
df2 = df2[df2.Age_group!='18+  years']
```

```python
df2.Age_group.unique()
```

```python
df2.sort_values(by=['Country','Year'],ignore_index=True, inplace=True)
```

```python
df2.head(20)
```

```python
df3 = pd.DataFrame(df2.groupby(['Country','Year','Gender'])['bmi'].mean())
type(df3)
# df3
```

```python
df3
```

```python
male_data = df3.query('Gender == "Male"')
male_data = male_data.sort_values(by=['Country', 'Year'],ignore_index=True)
male_data
```

```python
female_data = df3.query('Gender == "Female"')
female_data= female_data.sort_values(by=['Country', 'Year'],ignore_index=True)

female_data

```

```python
both_data = df3.query('Gender == "Both sexes"')
both_data= both_data.sort_values(by=['Country', 'Year'],ignore_index=True)

both_data
```

```python
df4 = df2[['Country','Year']]
df4.drop_duplicates(inplace=True)
```

```python
df4.shape
```

```python
df4.sort_values(by=['Country','Year'],ignore_index=True, inplace=True)
```

```python
df4['Male_bmi'] = male_data
df4['Female_bmi'] = female_data
df4['bmi'] = both_data
df4
```

```python
df4.info()
```

```python
df4.isna().sum()
```

```python
df4.to_csv("bmi_genderwise.csv")
```
