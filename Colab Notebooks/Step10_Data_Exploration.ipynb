```python
v1= df['percentage expenditure']
v2 = df['infant deaths']
```

```python
corr_coeff, p_value = kendalltau(v1, v2)

print("Kendall correlation coefficient:", corr_coeff)
print("P-value:", p_value)
```

**Kendall** correlation coefficient: The Kendall correlation coefficient also measures the monotonic relationship between two variables. It is less sensitive to outliers than Spearman correlation coefficient, but it is also less efficient. You can use Kendall correlation coefficient when you want to analyze the strength and direction of a monotonic relationship between two variables, especially when there are outliers in the data.

```python
c = [ 'Country', 'Year', 'Status', 'Life expectancy',
       'Adult Mortality', 'infant deaths', 'Alcohol', 'percentage expenditure',
       'under-five deaths', 'Total expenditure', 'HIV/AIDS',
       'thinness 1-19 years', 'Income composition of resources', 'Schooling',
       'Male_suicide_rate', 'Female_suicide_rate', 'suicide_rate',
       'Male_homicides', 'Female_homicides', 'Homicides', 'Male_bmi',
       'Female_bmi', 'bmi', 'Sanitation', 'GDP', 'Population']
kendall_corr_matrix = df[c].corr(method='kendall')
fig, ax = plt.subplots(figsize=(20,20)) 
sns.heatmap(kendall_corr_matrix, annot=True, vmin=-1, cmap='BrBG',ax=ax)
plt.show()
```
